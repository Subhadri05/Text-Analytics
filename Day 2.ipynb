```python
# ready many
cl = MongoClient()
db1 = cl.video
mov = db1.movieDetails
for x in mov.find({'year':1986}):
    print(x)
```

```python
## EXERCISE: Display the 'title' and 'year' of all the movies of the year 2000.

for y in mov.find({'year':2000}, {'_id':0, 'title':1, 'year':1}):
    print(y)
```

```python
## WAP to display only two fields 'emoji' and 'name'.

emojis_db = cl['emojis_db']
emo = emojis_db['emoji']
for i in emo.find({}, {'_id':0, 'emoji':1, 'name':1}):
    print(i)
```

```python
'''import pprint
qry = [{'$group': {'_id':'$year', 'runtime': {'$max':'$runtime'}}}]
for i in mov.aggregate(qry):
    pprint.pprint(i)'''

for i in mov.aggregate([{'$group': {'_id':'$year', 'runtime': {'$max':'$runtime'}}}]):
    print(i)
```

```python
# Print only the 4th and 5th document

for i in mov.find().limit(2).skip(3):
    print(i)
```

```python
## Movie names in which Dennis Hopper has acted.

for i in mov.find({'actors':'Dennis Hopper'}, {'_id':0, 'title':1}):
    print(i)
```

```python
## What are the different years present?

for i in mov.distinct('year'):
    print(i)
```

```python
## Each year, how many movy are there?

for i in mov.aggregate([{'$group': {'_id':'$year', 'count': {'$sum':1}}}, {'$sort':{'_id':1}}]):
    print(i)
```

```python
for i in mov.find({}, {'_id':0, 'imdb.votes':1}):
    print(i)
```

```python
## Movie names released after 1990.

for i in mov.find({'year':{'$gt':1990}}, {'_id':0, 'title':1}):
    print(i)
```
